# Get the figures for all countries

## Import useful library

```python
from bs4 import BeautifulSoup
import pandas
import time
import random
import requests
import pickle
import tqdm
```

### Create a file with the isocode of all the countries, with a file text

```python
with open ('iso_country_html.txt','r') as fp :
    html = fp.read()
    soup = BeautifulSoup(html,'html5lib')
    
trs = soup.find_all('tr')

country_dic = {}
country_isocode = []
for tr in trs:
    values = [td.string for td in tr.find_all ('td')]
    if values[0] is not None and len(values[0])>1:
        country_dic[values[0]]=values[3]
        country_isocode.append(values[3])
```

### Define some variables

### Request an API on World Bank to get the figures

```python
Index = range(1,250)
table = pandas.DataFrame(index=Index, columns=Columns)
table['Country Name']=country_dic
```

### Request an API on World Bank to get the figures

```python
def get_url(country_code, indicator):
    head = 'http://api.worldbank.org/v2/country/'
    middle = '/indicator/'
    tail = '?format=json'
    url = head + country_code + middle + indicator + tail
    r = requests.get(url)
    try:
        data = r.json()
        if data[1] is None:
            return None
        else:
            for i in data[1]:
                if i['value'] is None:
                    continue
                else:
                    return i['value']
    except KeyError:
        return None
    except IndexError:
        return None
```

```python
def get_value_on_WB(indicator, column):
    try:        
        for i, country in table['Country Name'].iteritems():
            country_code = country_dic[country]
            value = get_url(country_code, indicator)
            if value is not None:
                table.at[i, column] = value
            print(country,'done...')
            time.sleep(0.5)
        table[[i, column]]
    except KeyError:
        return None
    except IndexError:
        return None
```

```python
def fullfil_table(indic):
        for i in indic:
            get_value_on_WB(indic[i], i)

fullfil_table(indicator_list)
```
