## Import useful library

```python
from bs4 import BeautifulSoup
import pandas
import time
import random
import requests
import pickle
import tqdm
```

### Create a file with the isocode of all the countries, with a file text

```python
with open ('iso_country_html.txt','r') as fp :
    html = fp.read()
    soup = BeautifulSoup(html,'html5lib')
    
trs = soup.find_all('tr')

country_dic = {}
country_isocode = []
for tr in trs:
    values = [td.string for td in tr.find_all ('td')]
    if values[0] is not None and len(values[0])>1:
        country_dic[values[0]]=values[3]
        country_isocode.append(values[3])
```

### Define some variables

```python
indicator_list={
'IMF WB\nCustoms and other import duties (% of tax revenue) 2016':'GC.TAX.IMPT.ZS',
'WB IMF Non-performing loans to total gross loans ratio (%)':'FB.AST.NPER.ZS',
'UIC Rail freight (mn ton x km travelled)':'IS.RRS.GOOD.MT.K6',
'WB Household final consumption expenditure, PPP bn USD':'NE.CON.PRVT.PP.CD',
'WB UNESCO R&D Expenditure % of GDP':'GB.XPD.RSDV.GD.ZS',
'WB UNESCO Researchers p. mn inhabitants':'SP.POP.SCIE.RD.P6',
'IFDI Net Inflows (% of GDP)':'BX.KLT.DINV.WD.GD.ZS',
'WHO Current Health Expenditure p.c., PPP USD':'SH.XPD.CHEX.PP.CD',
'WB Research Expenditure % of GDP':'GB.XPD.RSDV.GD.ZS',
'WIPO Trademark applications':'IP.TMK.RSCT',
'WIPO Patent Applications':'IP.PAT.RESD'
}
```

### Request an API on World Bank to get the figures

```python
Index = range(1,250)
table = pandas.DataFrame(index=Index, columns=Columns)
table['Country Name']=country_dic
```

### Request an API on World Bank to get the figures

```python
def get_url(country_code, indicator):
    head = 'http://api.worldbank.org/v2/country/'
    middle = '/indicator/'
    tail = '?format=json'
    url = head + country_code + middle + indicator + tail
    r = requests.get(url)
    try:
        data = r.json()
        if data[1] is None:
            return None
        else:
            for i in data[1]:
                if i['value'] is None:
                    continue
                else:
                    return i['value']
    except KeyError:
        return None
    except IndexError:
        return None
```

```python
def get_value_on_WB(indicator, column):
    for i, country in table['Country Name'].iteritems():
        country_code = country_dic[country]
        value = get_url(country_code, indicator)
        if value is not None:
            table.at[i, column] = value
        print(country,'done...')
        time.sleep(0.5)
    table[[i, column]]
```
